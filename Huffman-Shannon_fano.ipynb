Huffman-Shannon_fano

```python
import math

L = 0
hs = 0
p = []
lk = []

# Number of samples
n = int(input("Enter the number of Samples : "))

# Input probabilities
for i in range(n):
    prob = float(input(f"Enter the probability of sample values {i+1}: "))
    p.append(prob)

# Input codeword lengths
for i in range(n):
    length = float(input(f"Enter the length of the sample values {i+1}: "))
    lk.append(length)

# Average Codeword Length
for i in range(n):
    L += p[i] * lk[i]

# Entropy
for i in range(n):
    if p[i] != 0:
        hs += p[i] * math.log(1 / p[i], 2)

hs = round(hs, 3)
L = round(L, 3)

# Efficiency
eff = round(hs / L, 3)

# Redundancy
red = round(1 - eff, 3)

# Variance
var = 0
for i in range(n):
    var += p[i] * (lk[i] - L) ** 2

var = round(var, 3)

# Output
print(f"Average Codeword Length is : {L}")
print(f"Entropy is : {hs}")
print(f"Efficiency is : {eff}")
print(f"Redudancy is : {red}")
print(f"Variance is : {var}")
```
